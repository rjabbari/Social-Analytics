```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
# Dependencies
import json
import tweepy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from config import consumer_key, consumer_secret, access_token, access_token_secret
```

```python
# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
# Target User
target_user_bbc = "BBCBreaking"

# Counter
counter = 1

# Variables for holding sentiments
bbc_sentiments = []

# Variable for max_id
oldest_tweet = None

# Loop through 5 pages of tweets (total 100 tweets)
for x in range(5):

    # Get all tweets from home feed
    public_tweets = api.user_timeline(target_user_bbc, max_id = oldest_tweet)

    # Loop through all tweets 
    for tweet in public_tweets:
 
        # Run Vader Analysis on each tweet
        results = analyzer.polarity_scores(tweet["text"])
        
        compound = results["compound"]
       
        tweets_ago = counter
        
        # Get Tweet ID, subtract 1, and assign to oldest_tweet
        oldest_tweet = tweet['id'] - 1
        
           # Add sentiments for each tweet into a list
        bbc_sentiments.append({"BBC":compound},)
        
        # Add to counter 
        counter += 1        
```

```python
bbc_df = pd.DataFrame(bbc_sentiments)
```

```python
# Target User
target_user_cbs = "CBS"

# Counter
counter = 1

# Variables for holding sentiments
cbs_sentiments = []

# Variable for max_id
oldest_tweet = None

# Loop through 5 pages of tweets (total 100 tweets)
for x in range(5):

    # Get all tweets from home feed
    public_tweets = api.user_timeline(target_user_cbs, max_id = oldest_tweet)

    # Loop through all tweets 
    for tweet in public_tweets:

        
        # Run Vader Analysis on each tweet
        results = analyzer.polarity_scores(tweet["text"])
        compound = results["compound"]
        
        tweets_ago = counter
        
        # Get Tweet ID, subtract 1, and assign to oldest_tweet
        oldest_tweet = tweet['id'] - 1
        
        # Add sentiments for each tweet into a list
        cbs_sentiments.append({"CBS":compound,})
                                  
        # Add to counter 
        counter += 1
```

```python
merge_data = bbc_df.join(CNN_df.join(fox_df.join(nyt_df.join(CBS_df, how='outer'))))
df = pd.DataFrame(merge_data)
df = df.rename_axis('Tweets', axis='columns')
df
```

```python
ax1 = df.reset_index().plot.scatter(x='index', y='BBC', c='DarkBlue', label='BBC');
ax2 = df.reset_index().plot.scatter(x='index', y='CNN', c='DarkRed', label='CNN', ax = ax1);
ax3 = df.reset_index().plot.scatter(x='index', y='Fox', c='DarkGreen', label='Fox', ax = ax2);
ax4 = df.reset_index().plot.scatter(x='index', y='NYT', c='DarkOrange', label='NYT', ax = ax3);
ax5 = df.reset_index().plot.scatter(x='index', y='CBS', c='Black', label='CBS', ax = ax4);

print(ax1 == ax2 == ax3 == ax4 == ax5) 


plt.title("Tweeter Sentiment by Network")
plt.ylabel("Sentiment")
plt.xlabel("Number of Tweets")
plt.xlim((0,100))
plt.grid(True)

# Create a legend
lgnd = plt.legend(fontsize="small", mode="Expanded", 
                  numpoints=1, scatterpoints=1, 
                  loc="best", title="Network", 
                  labelspacing=0.5)
lgnd.legendHandles[0]._sizes = [30]

# Save Figure
plt.savefig("bubble.png")
```

```python
compound = [df["BBC"].mean(), df["CNN"].mean(), df["Fox"].mean(), df["NYT"].mean(), df["CBS"].mean()]
```

```python
networks = ["BBC", "CNN", "Fox", "NYT", "CBS"]
x_axis = np.arange(len(compound))
```

```python
plt.bar(x_axis, compound)
plt.xticks(x_axis, networks)
plt.ylabel("Sentiment")
plt.title("Tweet Sentiment by News Organization")
plt.tight_layout()
plt.show()

plt.savefig("bar.png")
```
